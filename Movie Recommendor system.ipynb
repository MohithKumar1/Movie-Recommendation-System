What is the most frequent genre in the dataset ?

```python
data['Genre'] = data['Genre'].str.split(', ')
```

```python
pd.Series(data['Genre'].sum()).unique()
```

```python
unique_genres = pd.Series(data['Genre'].sum()).unique()

for genre in unique_genres:
    data[genre] = 0
    
    for i, row in data.iterrows():
        
        if genre in row['Genre']:
            data.at[i,genre] += 1
```

```python
pd.options.display.max_columns = None
```

```python
data.head()
```

```python
sorted_genre = data[unique_genres].sum().sort_values(ascending = True)
```

```python
sns.barplot(x = sorted_genre.index , y = sorted_genre.values , palette = 'PuBu')

plt.xticks(rotation = 90 )
plt.show()
```

What are the highest-rated movies in the dataset?

```python
Appropirate_vote_count = 3000
higest_rated_movies = data.sort_values('Vote_Average' , ascending = False)

highest_rated_movies = higest_rated_movies[higest_rated_movies['Vote_Count'] >= Appropirate_vote_count].head()
higest_rated_movies
```

```python
plt.bar( x = highest_rated_movies.Title ,  height =highest_rated_movies.Vote_Average.values )
plt.xticks(rotation = 90)
plt.show()
```

Which year Movies were most produced ?

```python
data['Year'] = data['Release_Date'].dt.year
movies_per_year = data.groupby('Year').count()['Title']
```

```python
movies_per_year.plot(kind = 'line', figsize = (10,5))

plt.title('Number of Movies per Year')
plt.ylabel("Number of Movies ")
plt.show()
```

```python
most_movies_year = movies_per_year.idxmax()
most_movies_year
```

During which month is the highest Movie production observed in the dataset?
